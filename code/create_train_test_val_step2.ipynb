```python
""" Import the necessary libraries """
import pandas as pd
from numpy import asarray
from sklearn.preprocessing import OneHotEncoder, StandardScaler, label_binarize
from statistics import mean 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
import numpy as np
import os
import shutil
```

```python
"""Read the test cohort indices from test_id.txt file"""

path = './dataset-splitting/test_id.txt'
# opening the file in read mode
my_file = open(path, "r")
  
# reading the file
data = my_file.read()
  
# replacing end splitting the text 
# when newline ('\n') is seen.
test_id = data.split("\n")
test_id = [int(x) for x in test_id]
print(test_id)
my_file.close()
```

```python
""" Create the test folder with WSIs with IDs from the test  """
lst = list(range(1,1058+1))

data_path = './dataset/original_WSI_patches'
data_dir = os.listdir(data_path)
pd_data_dir = pd.DataFrame(data_dir)
print(pd_data_dir)
# for each folder in the directory
for data_dir in test_id:
    print(data_dir)
    shutil.copytree(data_path + '/' + str(data_dir), './dataset/test' + '/' + str(data_dir))
```

```python
"""Read the train cohort indices from train_id.txt file"""

path = './dataset-splitting/train_id.txt'
# opening the file in read mode
my_file = open(path, "r")
  
# reading the file
data = my_file.read()
  
# replacing end splitting the text 
# when newline ('\n') is seen.

train_id = data.split("\n")
train_id = [int(x) for x in train_id]
print(train_id)
my_file.close()
```

```python
faulty_id = [125, 346, 572, 732, 870]
for i in train_id:
    if i in faulty_id:
        train_id.remove(i)
len(train_id)
```

```python
"""Create the train folder with WSIs with IDs from the train index file """

lst = list(range(1,1058+1))

data_path = './dataset/original_WSI_patches'
data_dir = os.listdir(data_path)
pd_data_dir = pd.DataFrame(data_dir)
#print(pd_data_dir)
train_data_dir = []
# for each folder in the directory
for data_dir in train_id:
    train_data_dir.append(data_dir)
    print(len(train_data_dir))
    shutil.copytree(data_path + '/' + str(data_dir), './dataset/train' + '/' + str(data_dir))
```
